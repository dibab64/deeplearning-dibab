```python
import tensorflow_datasets as tfds
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
ds, info = tfds.load('voc', split='train', shuffle_files=True, with_info=True)
```

```python
ds = ds.map(lambda x: x['image'], num_parallel_calls=tf.data.AUTOTUNE)
```

```python
tfds.visualization.show_examples(
    ds,
    ds_info=info,
    rows=1
)
```
